### chisquare test

```python
dataset_table=pd.crosstab(data1['sex'],data1['smoker'])
dataset_table
```

```python
observed_values=dataset_table.values
```

```python
observed_values
```

```python
expected_values=stats.chi2_contingency(dataset_table)
```

```python
expected_values
```

```python
expected_values=expected_values[3]
```

```python
expected_values
```

```python
no_of_rows=len(dataset_table.iloc[0:2,0])
no_of_columns=len(dataset_table.iloc[0,0:2])
ddof=(no_of_rows-1)*(no_of_columns-1)
print(ddof)
alpha=0.05
```

```python
from scipy.stats import chi2
chi_square=sum([(o-e)**2/e for o,e in zip(observed_values,expected_values)])
chisquare_statistic=chi_square[0]+chi_square[1]
print(chi_square)
print(chisquare_statistic)
```

```python
p_value=chi2.cdf(x=chisquare_statistic,df=ddof)
print(p_value)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.ensemble import RandomForestRegressor
```

### Linear Regression Model

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=0)
lin_reg=LinearRegression()
lin_reg.fit(xtrain,ytrain)
ypred=lin_reg.predict(xtest)
```

```python
print(r2_score(ytest,ypred))
```

### Polynomial Linear Regression

```python
x=data2.drop(columns=['charges','region'],axis=1)
y=data2.charges

quad=PolynomialFeatures(degree=2)
x_quad=quad.fit_transform(x)

xtrain,xtest,ytrain,ytest=train_test_split(x_quad,y,random_state=0)
plr=LinearRegression()
plr.fit(xtrain,ytrain)
ypred=plr.predict(xtest)
print(ypred)
```

```python
print(r2_score(ytest,ypred))
```

### Random Forest Regressor

```python
random_forest=RandomForestRegressor(n_estimators=120,criterion='mse',random_state=1)
random_forest.fit(xtrain,ytrain)
ypred=random_forest.predict(xtest)
print(r2_score(ytest,ypred)*100)
```

```python
r2score=[]
for i in range(30,100):
    random_forest=RandomForestRegressor(n_estimators=i,n_jobs=-1,criterion='mse',random_state=1)
    random_forest.fit(xtrain,ytrain)
    ypred=random_forest.predict(xtest)
    r2=r2_score(ytest,ypred)*100
    r2score.append(r2)

i=[i for i in range(30,100)]
sns.scatterplot(x=i,y=r2score,color='green')
    
```
