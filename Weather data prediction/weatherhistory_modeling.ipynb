### Now make Precipitation Type of weather data having snow and rain, make it as 0 and 1 Respectively.

```python
def value_change(var1):
    if var1=='snow':
        return 0
    else:
        return 1
    
# apply this value change function to change snow and rain in Precipitaton type in weather data
weather['Precip Type']=weather['Precip Type'].apply(value_change)
```

```python
weather['Precip Type'].unique()
```

```python
weather.head()
```

```python
# split the data to Independent and Dependent variables
x=weather.iloc[:,1:]
y=weather['Precip Type'].values.ravel()
print(x.head())
print(y)
```

### Modeling the data

```python
# Using sklearn to train and test the model, firstly have to split the data into training and testing
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2) # splitting the weather data of training size=80%, testing= 20%

print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
xtrain=scaler.fit_transform(xtrain)
xtest=scaler.fit_transform(xtest)
```

### Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression()
log_reg.fit(xtrain,ytrain)
ypred=log_reg.predict(xtest)
```

```python
ypred
```

```python
unseen=log_reg.predict([[8.288889,5.944444,0.83,14.1036,269.0,15.8263,1016.41]])
print(unseen)
```

```python
# creating Confusion matrix to identify the Actual and Predicted values
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,precision_score,recall_score,f1_score
cm_log=confusion_matrix(ytest,ypred)
sns.heatmap(cm_log,annot=True)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.show()
```

```python
accuracy_log=(accuracy_score(ytest,ypred)*100)
print("Accuracy Score of predicting the weather data is:",accuracy_log)
```

```python
# Classification report
report_log=classification_report(ytest,ypred)
print("Classification report of predicting the weather data is:\n",report_log)
```
